- Draw histograms and histogram variants for each feature or column.  (Swarm plot, kde plot, violin plot).

```python
# Took help from google and chatGPT to combine all dataframes and loop them 

import matplotlib.pyplot as plt
import seaborn as sns

dataframes = {"World Cup": df_worldcup, "Matches 1930-2022": df_matches, "FIFA Ranking": df_fifaranking}

def plot_charts(df, title):
    numeric_cols = df.select_dtypes(include=["number"]).columns  # Select numeric columns

    for col in numeric_cols:
        plt.figure(figsize=(15, 5))

        # Histogram
        plt.subplot(1, 4, 1)
        sns.histplot(df[col], bins=20, kde=True, color="blue")
        plt.title(f"{title} - Histogram of {col}")

        # KDE Plot
        plt.subplot(1, 4, 2)
        sns.kdeplot(df[col], fill=True, color="red")
        plt.title(f"{title} - KDE of {col}")

        # Violin Plot
        plt.subplot(1, 4, 3)
        sns.violinplot(y=df[col], color="green")
        plt.title(f"{title} - Violin Plot of {col}")

        # Swarm Plot (only if unique values are not too many)
        plt.subplot(1, 4, 4)
        if df[col].nunique() < 50:  # Swarm plot is best for limited unique values
            sns.swarmplot(y=df[col], color="purple")
            plt.title(f"{title} - Swarm Plot of {col}")
        else:
            plt.text(0.5, 0.5, "Too many unique values for Swarm", ha="center", va="center")
        
        plt.tight_layout()
        plt.show()

# Loop through all datasets and plot
for name, df in dataframes.items():
    print(f"### Plotting for {name} Dataset ###")
    plot_charts(df, name)
```

- Find means, medians, and modes.

```python
def compute_statistics(df, name):
    print(f"\n### {name} Dataset ###")
    numeric_cols = df.select_dtypes(include=["number"])

    mean_vals = numeric_cols.mean()
    median_vals = numeric_cols.median()
    mode_vals = numeric_cols.mode().iloc[0]  # Take the first mode if multiple exist

    # Print results
    stats_df = pd.DataFrame({"Mean": mean_vals, "Median": median_vals, "Mode": mode_vals})
    print(stats_df)

# Compute statistics for each dataset
compute_statistics(df_worldcup, "World Cup")
compute_statistics(df_matches, "Matches")
compute_statistics(df_fifaranking, "FIFA Ranking")
```
